```python
import numpy as np
import random
import pandas as pd
import matplotlib.pyplot as plt 

def coinFlip(size):
    xmin, xmax, dx =1, size, 1
    x = np.arange(xmin,xmax,dx) 
    flips = np.random.randint(0, 2, size=size) 
    
    return x, [flips[:i].mean() for i in x]

x,y = coinFlip(100)

#Drawing graph
plt.ylim([0,1])
plt.plot(x,y)
#table
i = 9

while i < len(y):
    print("#of coin tosses : %d  Var(Y) : %0.5f" % (i+1,y[i]))
    i += 10
    
```
